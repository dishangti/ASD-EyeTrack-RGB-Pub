```python
import sys
sys.path.append('..')
from eyelib import dataload
import numpy as np
import os
from datetime import datetime
```

```python
ASD_path = '../data/ASD/'
TD_path = '../data/TD/'

newdev_name = 'gts4lvwifichn'
olddev_name = 'HWBAH4'

ages = [[],[]]

for j,record_file in enumerate(os.listdir(ASD_path)):
    print('ASD', j)
    eye_coor, sti_type, exp_info, usr_info = dataload.read_record(ASD_path+record_file)
    if exp_info['testMode'] == 'video' and len(eye_coor) == 8:
        exp_time = datetime.strptime(usr_info['expDate'][:8],'%Y%m%d')
        birth_time = datetime.strptime(exp_info['birthday'], '%Y-%m-%d')
        age = (exp_time-birth_time).days / 365
        ages[0].append(age)

for j,record_file in enumerate(os.listdir(TD_path)):
    print('TD', j)
    eye_coor, sti_type, exp_info, usr_info = dataload.read_record(TD_path+record_file)
    if exp_info['testMode'] == 'video' and len(eye_coor) == 8:
        exp_time = datetime.strptime(usr_info['expDate'][:8],'%Y%m%d')
        birth_time = datetime.strptime(exp_info['birthday'], '%Y-%m-%d')
        age = (exp_time-birth_time).days / 365
        ages[1].append(age)

ages = np.array(ages, dtype=object)
```

```python
np.save('./results/video_ages.npy', ages)
```
